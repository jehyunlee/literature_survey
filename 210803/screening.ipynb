# group by year

```python
df_rt["PY"] = df_rt["PY"].astype(int)
df_rt["TC"] = df_rt["TC"].astype(int)
df_rt_yTCavg = df_rt[["PY", "TC"]].groupby("PY").mean()
df_rt_yTCavg.tail(5)
```

```python
df_rt[["PY", "TC"]].groupby("PY").count().shape
```

```python
df_count = df_rt[["PY", "TC"]].groupby("PY").count()

fig, axs = plt.subplots(ncols=2, figsize=(10, 5))
axs[0].bar(df_count.index, df_count["TC"], 
           fc="orange", ec='none', 
           label="No. of publication")

axs[1].scatter(df_rt["PY"], df_rt["TC"], s=20, alpha=0.3,
               c="b", ec="k", lw=0.5, 
               label="No. of citation")
axs[1].set_facecolor('none')
axs[1].plot(df_rt_yTCavg["TC"], color="skyblue", zorder=-1, label="avg. by year")
axs[1].set_ylim(0, )

axs[0].legend(fontsize="x-small")
h1, l1 = axs[1].get_legend_handles_labels()
axs[1].legend(handles=h1[::-1], labels=l1[::-1], fontsize="x-small")
fig.tight_layout()
fig.savefig("rooftop_pubcit.png", dpi=200)
```

```python
df_rt_yTChigh = pd.DataFrame(columns=df_rt.columns)

for py in range(1980, 2022):
    df_py = df_rt.query(f"PY == {py}")
    if df_py.shape[0] > 0:
        q1 = np.quantile(df_py["TC"], 0.25)
        q3 = np.quantile(df_py["TC"], 0.75)
        irq = q3-q1
        ub = q3 + 1.5*irq
        num_screened = df_py.query(f'(TC > {ub})').shape[0]
        print(f"{py}: {num_screened}/{df_py.shape[0]}")
        df_rt_yTChigh = df_rt_yTChigh.append(df_py.query(f"(TC > {ub})"))
    
df_rt_yTChigh.shape
```

# TC average

```python
df_rt["TC_yavg"] = df_rt[["PY", "TC"]].apply(lambda x: x["TC"]/df_rt_yTCavg.loc[x["PY"]], axis=1)
df_rt.sort_values("TC_yavg", ascending=False, inplace=True)
df_selected_TCavg = df_rt[["AU", "TI", "SO", "SO_abb", "DE", "TC", "TC_yavg", "PY", "DI"]].query("TC_yavg > 5")
```

```python
df_rt.shape
```

```python
df_rt_yTChigh["TC_yavg"] = df_rt_yTChigh.apply(lambda x: x["TC"]/df_rt_yTCavg.loc[x["PY"]], axis=1)
df_selected = df_rt_yTChigh[["AU", "TI", "SO", "SO_abb", "DE", "TC", "TC_yavg", "PY", "DI"]]
```

```python
df_selected_TCavg.shape
```

# 2021 papers: Nature, EES, Scientific Reports

```python
df_rt_2021 = df_rt.query('PY == 2021')
print(df_rt_2021.shape)
df_rt_2021.head(3)
```

```python
df_rt_highif = df_rt.loc[~df_rt["SO"].isna()]
df_selected_highif = df_rt_highif.loc[df_rt_highif["SO"].str.contains("Nature|Energy and Environmental Science|Energy and Buildings|Scientific Reports")]
df_selected_highif = df_selected_highif[["AU", "TI", "SO", "SO_abb", "DE", "TC", "TC_yavg", "PY", "DI"]]
print(df_selected_highif.shape)
```

# merging

```python
print(df_selected.shape[0])
print(df_selected_TCavg.shape[0])
print(df_selected_highif.shape[0])
```

```python
df_select = pd.concat([df_selected, df_selected_TCavg, df_selected_highif], axis=0)
df_select.drop_duplicates("DI", inplace=True)
df_select.sort_values("PY", ascending=False, inplace=True)
df_select["AU1"] = df_select["AU"].apply(lambda x: x[0] + " et al.")
df_select["AU_split"] = df_select["AU"].apply(lambda x: ", ".join(x))
df_select.drop(["AU"], axis=1, inplace=True)
df_select = df_select[["AU1", "AU_split", "TI", "SO", "SO_abb", "DE", "TC", "TC_yavg", "PY", "DI"]]
df_select.shape
```
