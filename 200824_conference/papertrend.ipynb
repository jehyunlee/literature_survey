```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("talk")
sns.set_style("whitegrid")
sns.set_palette("colorblind")
```

```python
years = list(range(1980, 2021))
pubs_renewable = [140, 188, 170, 168, 226, 167, 135, 119, 101, 122, 156, 190, 272, 279, 376, 284, 436, 384, 389, 401, 532, 777, 1008, 1427, 1689, 2074, 2414, 3452, 5256, 5668, 6058, 7671, 8972, 10101, 11362, 11656, 13625, 16024, 17999, 21173, 11916]
pubs_energy = [29414, 31114, 32080, 35418, 37805, 38063, 36245, 35611, 35488, 38651, 39565, 43342, 43732, 49231, 53558, 55163, 63704, 65268, 66079, 67623, 75352, 80184, 85835, 93293, 106902, 115845, 118515, 127306, 138525, 147398, 157844, 176052, 186907, 200520, 212975, 220282, 233189, 258110, 270728, 294170, 179887]
pubs_solar = [36, 52, 51, 61, 79, 60, 58, 38, 28, 36, 52, 63, 72, 78, 117, 78, 131, 102, 110, 103, 110, 153, 211, 300, 356, 418, 507, 703, 955, 1101, 1225, 1705, 2153, 2355, 2765, 2746, 3121, 3850, 4517, 5169, 3105]
pubs_pv = [4, 7, 2, 18, 16, 22, 14, 7, 5, 11, 8, 17, 15, 19, 49, 28, 46, 32, 45, 19, 40, 45, 64, 101, 106, 126, 164, 230, 308, 355, 377, 563, 666, 787, 942, 977, 1123, 1379, 1675, 1980, 1137]

pubs_es = [491, 542, 457, 488, 625, 558, 493, 509, 435, 478, 388, 520, 445, 529, 474, 516, 501, 532, 483, 552, 582, 712, 740, 975, 1217, 1273, 1553, 2028, 2537, 2078, 2525, 3602, 4546, 5425, 7052, 7926, 9616, 12067, 14173, 16808, 11942]
pubs_ess = [64, 57, 64, 47, 72, 56, 54, 54, 45, 64, 46, 76, 63, 56, 52, 78, 79, 91, 87, 116, 111, 118, 110, 124, 154, 195, 209, 282, 323, 488, 640, 810, 1031, 1281, 1655, 1851, 2512, 2954, 3719, 4685, 2832]

list(zip(years, pubs_pv, pubs_solar, pubs_renewable, pubs_energy, pubs_es, pubs_ess))
```

```python
df_pub = pd.DataFrame({"years":years, 
                       "pubs_pv":pubs_pv, 
                       "pubs_solar":pubs_solar, 
                       "pubs_renewable":pubs_renewable, 
                       "pubs_es":pubs_es,
                       "pubs_ess":pubs_ess,
                       "pubs_energy":pubs_energy})


df_pub["pv/solar"] = df_pub["pubs_pv"]/df_pub["pubs_solar"]*100
df_pub["solar/RE"] = df_pub["pubs_solar"]/df_pub["pubs_renewable"]*100
df_pub["RE/E"] = df_pub["pubs_renewable"]/df_pub["pubs_energy"]*100
df_pub["pv/RE"] = df_pub["pubs_pv"]/df_pub["pubs_renewable"]*100
df_pub["solar/E"] = df_pub["pubs_solar"]/df_pub["pubs_energy"]*100
df_pub["pv/E"] = df_pub["pubs_pv"]/df_pub["pubs_energy"]*100
df_pub["es/E"] = df_pub["pubs_es"]/df_pub["pubs_energy"]*100
df_pub["ess/E"] = df_pub["pubs_ess"]/df_pub["pubs_energy"]*100
df_pub.tail()
```

```python
df_pub[["years", "pubs_energy", "pubs_pv", "pubs_solar", "pubs_renewable", "pubs_es", "pubs_ess"]].to_csv("scopus_ess.csv")
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(13,5))

# 0. energy

ax[0].plot(df_pub["years"], df_pub["pubs_energy"], label='"energy"')
ax[0].plot(df_pub["years"], df_pub["pubs_renewable"], label='"renewable"AND"energy"')
ax[0].set_ylim(0,)
ax[0].set_xlim(1980, 2020)

yticks = ax[0].get_yticks()
ax[0].set_yticklabels([f"{int(y/1000)}K" for y in yticks])

ax[0].set_title("No. of Publications", y=1.05, # y=1.5, 
                fontdict={"fontweight":"bold", "color":"k"})
ax[0].text(1990, 20000, '"energy" AND "renewable"', 
           fontdict={"fontsize":14, "color": "C1"})
ax[0].text(1990, 80000, '"energy"', 
           fontdict={"fontsize":14, "color": "C0"})

# 1. portion

ax[1].plot(df_pub["years"], df_pub["RE/E"], c="k")
ax[1].set_ylim(0,)
ax[1].set_xlim(1980, 2020)
ax[1].set_title('Portion of "energy" (%)', y=1.05, # y=1.5, 
                fontdict={"fontweight":"bold", "color":"k"})
ax[1].text(2021, 7, '"energy" AND "renewable"',  
                fontdict={"color":"k", "fontsize":14})

ax[1].plot(df_pub["years"], df_pub["es/E"], c="r")
ax[1].text(2021, 6.5, '"energy storage"',  
           fontdict={"color":"r", "fontsize":14})


ax[1].plot(df_pub["years"], df_pub["ess/E"], c="green")
ax[1].text(2021, 1.8, '"energy storage system"',  
           fontdict={"color":"green", "fontsize":14})
                                   
plt.tight_layout()
plt.savefig("scopus_ess.png", dpi=300)
plt.show()
```
