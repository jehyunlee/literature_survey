```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
sns.set_context("talk")
```

```python
# good
ps = np.linspace(10, 20, 6)
x_noise = np.array([np.random.normal(loc=p, scale=0.2, size=30) for p in ps])
y_noise = np.array([np.random.normal(loc=p, scale=0.2, size=30) for p in ps])


```
